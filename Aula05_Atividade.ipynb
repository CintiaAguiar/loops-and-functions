# **Estrutura de repetição**

image.png

```python
for i in range (6):
  print("Regue a planta " + str(i))
```

```python
i = 0
while i < 6:
  print("Regue a planta " + str(i))
  i += 1
```

```python
for i in range(5):
  if i == 1 or i == 3 or i == 5:
    continue
  else:
    print("Regue os tomates de nº " + str(i))
```

```python
i = 0
while i < 5:
  tomates = str(i)
  print("Regue os tomates de nº " + tomates)
  i += 2
```

# **Função**

```python
def multiplicao (num1, num2):
  resultado = "A multiplicação de " + str(num1) + " x " + str(num2) + " = " + str(num1 * num2) + "\n"
  return resultado

resposta = multiplicao(5, 8)
print(resposta)
```

```python
def multiplicao(num1, num2):
  resultado = "Tabuada " + str(num1) + " x " + str(num2) + " = " + str(num1 * num2) + "\n"
  return resultado

for i in range(1, 11):
  resposta = multiplicao(9, i)
  print(resposta)
```
